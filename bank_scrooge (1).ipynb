 ### Анализ оттока клиентов. Доли отточных клиентов

Посчитаем средний процент оттока клиентов по банку

```python
average_percentage = round(100*df['churn'].mean(),2)
```

Построим столбчатые диаграммы доли отточных и лояльных клиентов по параметрам 'city','gender', 'equity', 'products', 'credit_card', 'last_activity'

```python
parameters=['city','gender', 'equity', 'products', 'credit_card', 'last_activity']

for i in range(len(parameters)):
    data= pd.pivot_table(df, index=parameters[i], columns='churn', values='user_id', aggfunc='count').reset_index()
    data.columns=[i, 'loyal', 'churn']
    data['percent_churn']=round(100*data['churn']/(data['churn']+data['loyal']),2)
    print(data)
    print('средний процент оттока клиентов по банку',average_percentage)
    ax = sns.barplot(x=i, y='percent_churn', data=data, estimator='mean') 
    plt.title(f'Доля отточных клиентов по {parameters[i]}')
    plt.ylabel('Процент оттока')
    plt.xlabel(f'параметр {parameters[i]}')
    plt.axhline(average_percentage, color='r', linestyle='--')
    plt.show()
```

Доля оттока клиентов в Ярославле (18,97) и Ростове (18,77) немного превышает средний процент оттока клиентов по банку, -18,2%.
Процент оттока у мужчин выше среднего значения _ 23,64%.
Самый высокий процент оттока - 53,85% - у клиентов с 9 баллами собстенности. Однако с учетом количества клиентов по данной категории и среднего процента оттока по банку, следует расмотреть выборки клиентов с 3, 4 и 5 баллами собственности. 
Доля отточных клиентов, которые пользуются двумя и более продуктами банка, выше среднего процента оттока клиентов.
Клиенты, которые пользуются кредитными картами, чаще становятся "отточными", 25,56% таких клиентов намереваются уйти из банка.
Интересно отметить, что активные в последнее время клиенты чаще принимают решение уйти из банка, процент оттока активных клиентов -24,41%.

## Статистический анализ

### Гипотеза различия дохода между теми клиентами, которые ушли и теми, которые остались

Сформулируем нулевую и альтернативную гипотезы:
Н0: средний доход отточных и лояльных клиентов равны.
Н1: средний доход отточных и лояльных клиентов не равны.

```python
sample_1=df[df['churn']==0]['est_salary']
sample_2=df[df['churn']==1]['est_salary']

alpha = 0.05 # уровень статистической значимости
# если p-value окажется меньше него, отвергнем гипотезу

results = st.ttest_ind(sample_1, sample_2,equal_var=False)

print('p-значение:', results.pvalue)

if results.pvalue < alpha:
    print('Отвергаем нулевую гипотезу')
else:
    print('Не получилось отвергнуть нулевую гипотезу') 
```

Проведенный t-test показал, что опровергнуть гипотезу о различии среднего дохода между клиентами, которые ушли и теми, которые остались, не удалось. 

### Гипотеза различия количества банковских продуктов между теми клиентами, которые ушли и теми, которые остались

Сформулируем нулевую и альтернативную гипотезы:
Н0: среднее количество банковских продуктов у отточных и лояльных клиентов равны.
Н1: среднее количество банковских продуктов у отточных и лояльных клиентов не равны.

```python
sample_1=df[df['churn']==0]['products']
sample_2=df[df['churn']==1]['products']

alpha = 0.05 # уровень статистической значимости
# если p-value окажется меньше него, отвергнем гипотезу

results = st.ttest_ind(sample_1, sample_2,equal_var=False)

print('p-значение:', results.pvalue)

if results.pvalue < alpha:
    print('Отвергаем нулевую гипотезу')
else:
    print('Не получилось отвергнуть нулевую гипотезу') 
```

Проведенный t-test показал, что нулевую гипотезу о различии среднего количества банковских продуктов у отточных и лояльных клиентов сдедует опровергнуть. Что подтверждает результаты исследовательского анализа.

### Гипотеза различия количества баллов собственности между теми клиентами, которые ушли и теми, которые остались

Сформулируем нулевую и альтернативную гипотезы:
Н0: среднее количество баллов собственности у отточных и лояльных клиентов равны.
Н1: среднее количество баллов собственности у отточных и лояльных клиентов не равны.

```python
sample_1=df[df['churn']==0]['equity']
sample_2=df[df['churn']==1]['equity']

alpha = 0.05 # уровень статистической значимости
# если p-value окажется меньше него, отвергнем гипотезу

results = st.ttest_ind(sample_1, sample_2,equal_var=False)

print('p-значение:', results.pvalue)

if results.pvalue < alpha:
    print('Отвергаем нулевую гипотезу')
else:
    print('Не получилось отвергнуть нулевую гипотезу') 
```

Проведенный t-test показал, что нулевую гипотезу о различии среднего количества баллов собственности у отточных и лояльных клиентов сдедует опровергнуть. Что подтверждает результаты исследовательского анализа.

### Гипотеза различия количества мужчин среди отточных и лояльных клиентов

Сформулируем нулевую и альтернативную гипотезы:
Н0: среднее количество мужчин среди отточных и лояльных клиентов равны.
Н1: среднее количество мужчин среди отточных и лояльных клиентов не равны.

```python
sample_1=df[df['churn']==0]['М']
sample_2=df[df['churn']==1]['М']

alpha = 0.05 # уровень статистической значимости
# если p-value окажется меньше него, отвергнем гипотезу

results = st.ttest_ind(sample_1, sample_2,equal_var=False)

print('p-значение:', results.pvalue)

if results.pvalue < alpha:
    print('Отвергаем нулевую гипотезу')
else:
    print('Не получилось отвергнуть нулевую гипотезу') 
```

Как и по итогам исследовательского анализа, проведенный тест доказал, что среднее количество мужчин среди отточных и лояльных клиентов не равны

### Гипотеза различия количества кредитных карт между теми клиентами, которые ушли и теми, которые остались

Сформулируем нулевую и альтернативную гипотезы:
Н0: среднее количество кредитных карт среди отточных и лояльных клиентов равны.
Н1: среднее количество кредитных карт среди отточных и лояльных клиентов не равны.

```python
sample_1=df[df['churn']==0]['credit_card']
sample_2=df[df['churn']==1]['credit_card']

alpha = 0.05 # уровень статистической значимости
# если p-value окажется меньше него, отвергнем гипотезу

results = st.ttest_ind(sample_1, sample_2,equal_var=False)

print('p-значение:', results.pvalue)

if results.pvalue < alpha:
    print('Отвергаем нулевую гипотезу')
else:
    print('Не получилось отвергнуть нулевую гипотезу') 
```

Проведенный тест подтверждает итоги исследовательского анализа, что среднее количество кредитных карт среди отточных и лояльных клиентов не равны.
